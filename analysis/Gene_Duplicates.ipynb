# Functions

## FPKM

```python
def fpkm(gene_count, fragment_length):
    ''' Fragments Per Kilobase of transcript per Million mapped reads '''
    gene_count = gene_count.join([fragment_length])
    gene_count.index.names = ['index']
    gene_count = gene_count.reset_index()

    gene_count.loc['sum'] = gene_count.sum(axis = 0)
    gene_count[gene_count.columns[1:]] = gene_count[gene_count.columns[1:]].astype(float)

    ## MULTIPLY BY 10^(9)
    gene_count.loc['sum', gene_count.columns[1:-1]] = pow(10,9)/gene_count.loc['sum'][gene_count.columns[1:-1]]
    FPKM = gene_count[:-1][gene_count.columns[1:-1]]

    ## DIVIDE BY TOTAL_READS
    for i in FPKM.columns:
        FPKM[i] = FPKM[i]*gene_count.at['sum', i]

    ## DIVIDE BY GENE_LENGTH
    for i in list(FPKM.index.values):
        FPKM.iloc[i] = FPKM.iloc[i]/gene_count.at[i, fragment_length.columns[0]]

    FPKM = pd.concat([FPKM, gene_count['index'][:-1]], axis = 1)

    # REDO COLUMN LABELS
    FPKM = FPKM.set_index('index')
    columns = FPKM.columns.str.replace('_','-', regex = True)
    FPKM.columns = columns
    return FPKM
```

## Identifying Paralogous Best Hits

```python
def paralogous_best_hit(blast_file1, blast_file2, name):
    
    ''' Finding top between species paralogous best hits'''
    
    file = pd.read_csv('/research/projects/chlamydomonas/NIexpression/data/orthologs/' + blast_file1, delimiter = '\t', usecols = [i for i in range(4)] + [12], names = ['sample1', 'sample2', 'percent_id', 'alignment_length', 'query_cov'])
    file1 = pd.read_csv('/research/projects/chlamydomonas/NIexpression/data/orthologs/' + blast_file2, delimiter = '\t', usecols = [i for i in range(4)] + [12], names = ['sample1', 'sample2', 'percent_id', 'alignment_length', 'query_cov'])
    
    #### Naming samples ####
    sample1 = name.split('_')[0]
    sample2 = name.split('_')[1]
    
    blasts = pd.read_csv('/research/projects/chlamydomonas/NIexpression/data/orthologs/BLASTRBH_orthologous_pairs.txt', delimiter = '\t', low_memory=True)
    blasts.columns = blasts.columns.str.replace('MALA', 'CMAL').str.replace('JG05', 'JG5-1')
    blasts = blasts[[sample1, sample2]].dropna()
    
    #### Formatting ####
    file['sample1'] = file['sample1'].str.split('_4532', expand = True)[0].str.split('.t', expand = True)[0]
    file['sample2'] = file['sample2'].str.split('_4532', expand = True)[0].str.split('.t', expand = True)[0]
        
    file1['sample1'] = file1['sample1'].str.split('_4532', expand = True)[0].str.split('.t', expand = True)[0]
    file1['sample2'] = file1['sample2'].str.split('_4532', expand = True)[0].str.split('.t', expand = True)[0]
    
    #### Finding duplicate genes ####
    file_duplicated = file.loc[file.duplicated(subset = 'sample2')]['sample2'].unique()
    file1_duplicated = file1.loc[file1.duplicated(subset = 'sample2')]['sample2'].unique()
     
    #### Putative duplicates ####
    file_dups = file.loc[file['sample2'].isin(file_duplicated)]
    file_dups = file_dups.loc[(file_dups['sample1'].isin(blasts[sample1])) | (file_dups['sample2'].isin(blasts[sample2]))]
    file_dups['sample1/sample2'] = file_dups['sample1'] + '_' + file_dups['sample2']
    blasts['CINC/CREN'] = blasts[sample1] + '_' + blasts[sample2]
    file_dups = file_dups[~file_dups['sample1/sample2'].isin(blasts['CINC/CREN'])]
    file_dups = file_dups.loc[(file_dups['query_cov'] >= 20) & (file_dups['percent_id'] >= 50)] ## kept all putative duplicates that have a query coverage over 20 and a percent identity over 50%
    
    file1_dups = file1.loc[file1['sample2'].isin(file1_duplicated)]
    file1_dups = file1_dups.loc[(file1_dups['sample1'].isin(blasts[sample2])) | (file1_dups['sample2'].isin(blasts[sample1]))]
    file1_dups['sample1/sample2'] = file1_dups['sample1'] + '_' + file1_dups['sample2']
    blasts['CREN/CINC'] = blasts[sample2] + '_' + blasts[sample1]
    file1_dups = file1_dups[~file1_dups['sample1/sample2'].isin(blasts['CREN/CINC'])]
    file1_dups = file1_dups.loc[(file1_dups['query_cov'] >= 20) & (file1_dups['percent_id'] >= 50)] ## kept all putative duplicates that have a query coverage over 20 and a percent identity over 50%
    file1_dups = file1_dups[['sample2', 'sample1', 'percent_id', 'alignment_length', 'query_cov']]
    file1_dups.columns = ['sample1', 'sample2', 'percent_id', 'alignment_length', 'query_cov']
    
    dups = pd.concat([file_dups, file1_dups], axis = 0)[['sample1', 'sample2']]
    dups.columns = [sample1, sample2]
    
    return dups
```

## Identifying Putative Paralogs

```python
def potential_paralogs(species_file, GFF_file):
    
    ''' Identifying all gene duplicates and classifying them according to duplication type'''
    
    species_file['sample1'] = species_file['sample1'].str.split('.t', expand = True)[0].str.split('_4532', expand = True)[0]
    species_file['sample2'] = species_file['sample2'].str.split('.t', expand = True)[0].str.split('_4532', expand = True)[0]
    
    ## Appending all non-identical gene duplicates ##
    index = []
    for i in species_file.index:
        if species_file.loc[i, 'sample1'] != species_file.loc[i, 'sample2']:
            index.append(i)
    species_file = species_file.loc[index]
    species_file = species_file.loc[(species_file['query_cov'] >= 20) & (species_file['percent_id'] >= 90)] ## kept all putative duplicates that have a query coverage over 20 and a percent identity  over 90%
    species_file.drop_duplicates(subset = ['sample1', 'sample2'], keep = 'first', inplace = True)
    
    #### Editing ####
    for i in species_file['sample1'].unique():
        section = species_file.loc[species_file['sample2'] == i]
        species_file.loc[section.index, 'sample1'] = section['sample2']
        species_file.loc[section.index, 'sample2'] = section['sample1']
    
    #### Removing duplicate blast hits ####
    species_file.drop_duplicates(subset = ['sample1', 'sample2'], keep = 'first', inplace = True)
    
    #### Finding the number of genes between putative duplicates ####
    for i in species_file.index:
        index1 = GFF_file.loc[GFF_file['Parent'] == species_file.loc[i, 'sample1']].index[0]
        index2 = GFF_file.loc[GFF_file['Parent'] == species_file.loc[i, 'sample2']].index[0]
        species_file.loc[i, 'genes_between'] = abs(index1 - index2) - 1
    
    #### Defining duplication types ####
    species_file['dup_type'] = 'dispersed'
    species_file.loc[species_file.loc[species_file['genes_between'] == 0].index, 'dup_type'] = 'tandem'             
    species_file.loc[species_file.loc[(species_file['genes_between'] != 0) & (species_file['genes_between'] <= 10)].index, 'dup_type'] = 'proximal' 
    
    for i in species_file['sample2'].unique():
        section = species_file.loc[species_file['sample2'] == i]
        if len(section) >= 2:
            for a in [[list(section.index)[e], list(section.index)[e+1]] for e in range(len(section.index) - 1)]:
                section1 = section.loc[a]
                min_number = section1['genes_between'].min()
                max_number = section1['genes_between'].max()
                max_index = GFF_file.loc[GFF_file['Parent'].isin(section1['sample1'])].index.max()
                min_index = GFF_file.loc[GFF_file['Parent'].isin(section1['sample1'])].index.min()
    
                if (max_number - min_number == 1) & (max_index - min_index == 1):
                    species_file.loc[section.index, 'dup_type'] = 'distant_tandem'
            
    for i in species_file['sample1'].unique():
        section = species_file.loc[species_file['sample1'] == i]
        if len(section) >= 2:
            for a in [[list(section.index)[e], list(section.index)[e+1]] for e in range(len(section.index) - 1)]:
                section1 = section.loc[a]
                min_number = section1['genes_between'].min()
                max_number = section1['genes_between'].max()
                max_index = GFF_file.loc[GFF_file['Parent'].isin(section1['sample2'])].index.max()
                min_index = GFF_file.loc[GFF_file['Parent'].isin(section1['sample2'])].index.min()
    
                if (max_number - min_number == 1) & (max_index - min_index == 1):
                    species_file.loc[section.index, 'dup_type'] = 'distant_tandem' 
    return species_file
```

## Identifying Species-specific Gene Duplicates

```python
def species_specific_duplicates(paralogs, dups_type_sp1, dups_type_sp2, fpkm_sp1, fpkm_sp2, species1, species2):
    """ This function provides the number of gene duplicates specific to each species.
    Species 1 and 2 should be a four-letter name in caps (e.g., CREN, CINC)"""
    
    blasts = pd.read_csv('/research/projects/chlamydomonas/NIexpression/data/orthologs/BLASTRBH_orthologous_pairs.txt', delimiter = '\t', low_memory=True)
    blasts.columns = blasts.columns.str.replace('MALA','CMAL').str.replace('JG05','JG5-1')
    blasts = blasts[[species1, species2]].dropna()
    
    Cren_newdups = pd.DataFrame()
    Cinc_newdups = pd.DataFrame()
    
    for i in paralogs.index:
        sp1 = paralogs.loc[i, species1]
        sp2 = paralogs.loc[i, species2]
        cinc = dups_type_sp1.loc[(dups_type_sp1['sample1'] == sp1) | (dups_type_sp1['sample2'] == sp1)]
        cren = dups_type_sp2.loc[(dups_type_sp2['sample1'] == sp2) | (dups_type_sp2['sample2'] == sp2)]
        
        if (len(cren) != len(cinc)):
            if (len(cinc) == 0) or (len(cren) == 0):
                Cren_newdups = pd.concat([Cren_newdups, cren])
                Cinc_newdups = pd.concat([Cinc_newdups, cinc])
            
    Cren_newdups.drop_duplicates(subset = ['sample1', 'sample2'], keep = 'first', inplace = True)
    Cinc_newdups.drop_duplicates(subset = ['sample1', 'sample2'], keep = 'first', inplace = True)
    
    #### Identifying orthologous - paralogous gene pairs ####
    for i in Cren_newdups.index:
        try:
            Cren_newdups.loc[i, species1 + '_ortholog'] = blasts.loc[blasts[species2].isin([Cren_newdups.loc[i, 'sample1']] + [Cren_newdups.loc[i, 'sample2']])][species1].values[0]
            Cren_newdups.loc[i, species2 + '_ortholog'] = blasts.loc[blasts[species1] == Cren_newdups.loc[i, species1 + '_ortholog']][species2].values[0]
        except: continue
    
    for i in Cinc_newdups.index:
        try:
            Cinc_newdups.loc[i, species2 + '_ortholog'] = blasts.loc[blasts[species1].isin([Cinc_newdups.loc[i, 'sample1']] + [Cinc_newdups.loc[i, 'sample2']])][species2].values[0]
            Cinc_newdups.loc[i, species1 + '_ortholog'] = blasts.loc[blasts[species2] == Cinc_newdups.loc[i, species2 + '_ortholog']][species1].values[0]
        except: continue
            
    Cren_newdups = Cren_newdups.dropna()
    Cinc_newdups = Cinc_newdups.dropna()
    
    #### Calculating expression divergence species 1 ####
    for index in Cinc_newdups.index:
        ed = []
        cr = Cinc_newdups.loc[index, species2 + '_ortholog']
        ci_orth = Cinc_newdups.loc[index, species1 + '_ortholog']
        ci_para = list(set(Cinc_newdups.loc[index, ['sample1', 'sample2']]).difference([Cinc_newdups.loc[index, species1 + '_ortholog']]))[0]
        Cinc_newdups.loc[index, species1 + '_paralog'] = ci_para
        
        cr2 = (fpkm_sp2.loc[cr][0] * (10**(6)))/(fpkm_sp2.sum(axis = 0)[0])
        ci_orth2 = (fpkm_sp1.loc[ci_orth][0] * (10**(6)))/(fpkm_sp1.sum(axis = 0)[0])
        ci_para2 = (fpkm_sp1.loc[ci_para][0] * (10**(6)))/(fpkm_sp1.sum(axis = 0)[0])
        
        Cinc_newdups.loc[index, 'TPM_' + species2 + '_ortholog'] = cr2 ## FPKM Normalized reads counts
        Cinc_newdups.loc[index, 'TPM_' + species1 + '_ortholog'] = ci_orth2
        Cinc_newdups.loc[index, 'TPM_' + species1 + '_paralog'] = ci_para2
    
        if 0 in (cr2, ci_orth2, ci_para2): 
            continue
        else:
            for i in [(cr2, ci_orth2), (cr2, ci_para2), (ci_orth2, ci_para2)]:
                ED = abs((i[0] - i[1])/(abs(i[0] + i[1]) + 2))
                ed.append(ED)
            Cinc_newdups.loc[index, 'ED:orth'] = ed[0]
            Cinc_newdups.loc[index, 'ED:paraBH'] = ed[1]
            Cinc_newdups.loc[index, 'ED:para'] = ed[2]
    
    #### Calculating expression divergence species 2 ####
    for index in Cren_newdups.index:
        ed = []
        ci = Cren_newdups.loc[index, species1 + '_ortholog']
        cr_orth = Cren_newdups.loc[index, species2 + '_ortholog']
        cr_para = list(set(Cren_newdups.loc[index, ['sample1', 'sample2']]).difference([Cren_newdups.loc[index, species2 + '_ortholog']]))[0]
        Cren_newdups.loc[index, species2 + '_paralog'] = cr_para
        
        ci2 = (fpkm_sp1.loc[ci][0] * (10**(6)))/(fpkm_sp1.sum(axis = 0)[0])
        cr_orth2 = (fpkm_sp2.loc[cr_orth][0] * (10**(6)))/(fpkm_sp2.sum(axis = 0)[0])
        cr_para2 = (fpkm_sp2.loc[cr_para][0] * (10**(6)))/(fpkm_sp2.sum(axis = 0)[0])
        
        Cren_newdups.loc[index, 'TPM_' + species1 + '_ortholog'] = ci2 ## FPKM Normalized reads counts
        Cren_newdups.loc[index, 'TPM_' + species2 + '_ortholog'] = cr_orth2
        Cren_newdups.loc[index, 'TPM_' + species2 + '_paralog'] = cr_para2
    
        if 0 in (ci2, cr_orth2, cr_para2):
            continue
        else:
            for i in [(ci2, cr_orth2), (ci2, cr_para2), (cr_orth2, cr_para2)]:
                ED = abs((i[0] - i[1])/(abs(i[0] + i[1]) + 2))
                ed.append(ED)
            Cren_newdups.loc[index, 'ED:orth'] = ed[0]
            Cren_newdups.loc[index, 'ED:paraBH'] = ed[1]
            Cren_newdups.loc[index, 'ED:para'] = ed[2]
            
    Cren_newdups = Cren_newdups.drop(columns = ['sample1', 'sample2']).dropna()
    Cinc_newdups = Cinc_newdups.drop(columns = ['sample1', 'sample2']).dropna()
        
    return Cinc_newdups, Cren_newdups
```

```python
def species_specific_duplicates2(paralogs, dups_type_sp1, dups_type_sp2, fpkm_sp1, fpkm_sp2, species1, species2):
    """ This function provides the number of gene duplicates specific to each species.
    Species 1 and 2 should be a four-letter name in caps (e.g., CREN, CINC)"""
    
    blasts = pd.read_csv('/research/projects/chlamydomonas/NIexpression/data/orthologs/BLASTRBH_orthologous_pairs.txt', delimiter = '\t', low_memory=True)
    blasts.columns = blasts.columns.str.replace('MALA','CMAL').str.replace('JG05','JG5-1')
    blasts = blasts[[species1, species2]].dropna()
    
    Cren_newdups = pd.DataFrame()
    Cinc_newdups = pd.DataFrame()
    
    for i in paralogs.index:
        sp1 = paralogs.loc[i, species1]
        sp2 = paralogs.loc[i, species2]
        cinc = dups_type_sp1.loc[(dups_type_sp1['sample1'] == sp1) | (dups_type_sp1['sample2'] == sp1)]
        cren = dups_type_sp2.loc[(dups_type_sp2['sample1'] == sp2) | (dups_type_sp2['sample2'] == sp2)]
        
        if (len(cren) != len(cinc)):
            if (len(cinc) == 0) or (len(cren) == 0):
                Cren_newdups = pd.concat([Cren_newdups, cren])
                Cinc_newdups = pd.concat([Cinc_newdups, cinc])
            
    Cren_newdups.drop_duplicates(subset = ['sample1', 'sample2'], keep = 'first', inplace = True)
    Cinc_newdups.drop_duplicates(subset = ['sample1', 'sample2'], keep = 'first', inplace = True)
    
    #### Identifying orthologous - paralogous gene pairs ####
    for i in Cren_newdups.index:
        try:
            Cren_newdups.loc[i, species1 + '_ortholog'] = blasts.loc[blasts[species2].isin([Cren_newdups.loc[i, 'sample1']] + [Cren_newdups.loc[i, 'sample2']])][species1].values[0]
            Cren_newdups.loc[i, species2 + '_ortholog'] = blasts.loc[blasts[species1] == Cren_newdups.loc[i, species1 + '_ortholog']][species2].values[0]
        except: continue
    
    for i in Cinc_newdups.index:
        try:
            Cinc_newdups.loc[i, species2 + '_ortholog'] = blasts.loc[blasts[species1].isin([Cinc_newdups.loc[i, 'sample1']] + [Cinc_newdups.loc[i, 'sample2']])][species2].values[0]
            Cinc_newdups.loc[i, species1 + '_ortholog'] = blasts.loc[blasts[species2] == Cinc_newdups.loc[i, species2 + '_ortholog']][species1].values[0]
        except: continue
            
    Cren_newdups = Cren_newdups.dropna()
    Cinc_newdups = Cinc_newdups.dropna()
    
    #### Calculating expression divergence species 1 ####
    for index in Cinc_newdups.index:
        ed = []
        cr = Cinc_newdups.loc[index, species2 + '_ortholog']
        ci_orth = Cinc_newdups.loc[index, species1 + '_ortholog']
        ci_para = list(set(Cinc_newdups.loc[index, ['sample1', 'sample2']]).difference([Cinc_newdups.loc[index, species1 + '_ortholog']]))[0]
        Cinc_newdups.loc[index, species1 + '_paralog'] = ci_para
        
        cr2 = (fpkm_sp2.loc[cr][0] * (10**(6)))/(fpkm_sp2.sum(axis = 0)[0])
        ci_orth2 = (fpkm_sp1.loc[ci_orth][0] * (10**(6)))/(fpkm_sp1.sum(axis = 0)[0])
        ci_para2 = (fpkm_sp1.loc[ci_para][0] * (10**(6)))/(fpkm_sp1.sum(axis = 0)[0])
        
        Cinc_newdups.loc[index, 'TPM_' + species2 + '_ortholog'] = cr2 ## FPKM Normalized reads counts
        Cinc_newdups.loc[index, 'TPM_' + species1 + '_ortholog'] = ci_orth2
        Cinc_newdups.loc[index, 'TPM_' + species1 + '_paralog'] = ci_para2
    
        for i in [(cr2, ci_orth2), (cr2, ci_para2), (ci_orth2, ci_para2)]:
            ED = abs((i[0] - i[1])/(abs(i[0] + i[1]) + 2))
            ed.append(ED)
        Cinc_newdups.loc[index, 'ED:orth'] = ed[0]
        Cinc_newdups.loc[index, 'ED:paraBH'] = ed[1]
        Cinc_newdups.loc[index, 'ED:para'] = ed[2]
    
    #### Calculating expression divergence species 2 ####
    for index in Cren_newdups.index:
        ed = []
        ci = Cren_newdups.loc[index, species1 + '_ortholog']
        cr_orth = Cren_newdups.loc[index, species2 + '_ortholog']
        cr_para = list(set(Cren_newdups.loc[index, ['sample1', 'sample2']]).difference([Cren_newdups.loc[index, species2 + '_ortholog']]))[0]
        Cren_newdups.loc[index, species2 + '_paralog'] = cr_para
        
        ci2 = (fpkm_sp1.loc[ci][0] * (10**(6)))/(fpkm_sp1.sum(axis = 0)[0])
        cr_orth2 = (fpkm_sp2.loc[cr_orth][0] * (10**(6)))/(fpkm_sp2.sum(axis = 0)[0])
        cr_para2 = (fpkm_sp2.loc[cr_para][0] * (10**(6)))/(fpkm_sp2.sum(axis = 0)[0])
        
        Cren_newdups.loc[index, 'TPM_' + species1 + '_ortholog'] = ci2 ## FPKM Normalized reads counts
        Cren_newdups.loc[index, 'TPM_' + species2 + '_ortholog'] = cr_orth2
        Cren_newdups.loc[index, 'TPM_' + species2 + '_paralog'] = cr_para2
    
        for i in [(ci2, cr_orth2), (ci2, cr_para2), (cr_orth2, cr_para2)]:
            ED = abs((i[0] - i[1])/(abs(i[0] + i[1]) + 2))
            ed.append(ED)
        Cren_newdups.loc[index, 'ED:orth'] = ed[0]
        Cren_newdups.loc[index, 'ED:paraBH'] = ed[1]
        Cren_newdups.loc[index, 'ED:para'] = ed[2]
            
    Cren_newdups = Cren_newdups.drop(columns = ['sample1', 'sample2']).dropna()
    Cinc_newdups = Cinc_newdups.drop(columns = ['sample1', 'sample2']).dropna()
        
    return Cinc_newdups, Cren_newdups
```
